$$\frac{dy}{dt}=y+t$$


```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

```python
n = 25 # numero de pontos

y = np.linspace (-2,2,n) #cria uma sequencia de numeros espaados de forma uniforme em que "n" define o numero de pontos dentro do intervalo
t = np.linspace (-2,2,n)

Y,T = np.meshgrid(y,t) #cria uma malha de pontos ao invs de apenas uma linha (como uma matrtiz)

f = y+T

angulo = np.arctan(f)
dy = np.cos(angulo)
dt = np.sin(angulo)

plt.quiver(Y,T,dy,dt)
```
